```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("All libraries imported successfully!")
```

```python
# load and read dataset

df=pd.read_csv(r"C:\Users\HP\Desktop\credit-card-fraud-detection\data\creditcard.csv")


# finding the shape of dataframe
print("shape :",df.shape)

# head() shows the first 5 rows of the dataset
print("\nFirst 5 rows:")
df.head()
```

```python
# Basic information about the dataset
print(" DATASET INFO")
df.info()
```

```python
# checking the dataset have missing value 
print("missing value")
print(df.isnull().sum())

# cheking # count, mean, min, max, 25%, 50%, 75%
print("statical info")
df.describe()
```

```python
# Count how many legitimate vs fraud transactions we have— shows our imbalance problem
print("CLASS DISTRIBUTION")
print(df['Class'].value_counts())

# Calculate fraud percentage
fraud_count = df['Class'].sum()
print("Fraud (Class 1):",fraud_count)
total_count = len(df)
fraud_percent = (fraud_count / total_count) * 100
print("Fraud percentage:",round(fraud_percent,4))
```

```python
# Visualise the class imbalance using a bar chart
# This makes it easy to see how extreme the imbalance is
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# chart 1 --count of each class--
counts = df['Class'].value_counts()
axes[0].bar(
    ['Legitimate (0)', 'Fraud (1)'],
    counts.values,
    color=['#2196F3', '#F44336'], # blue for legit, red for fraud
    edgecolor='white',
    linewidth=1.5

)
axes[0].set_title("Transaction Class Distribution",fontsize=14,fontweight='bold')
axes[0].set_ylabel('Count')

# Add count labels on top of each bar
for i, v in enumerate(counts.values):
    axes[0].text(
        i, v + 1000,
        f'{v:,}\n({v/len(df)*100:.2f}%)',
        ha='center',
        fontweight='bold'
    )



# ── Chart 2: Amount distribution by class ─────────────────
df[df['Class'] == 0]['Amount'].hist(
    bins=50, alpha=0.6,
    color='#2196F3',
    ax=axes[1],
    label='Legitimate'
)

df[df['Class'] == 1]['Amount'].hist(
    bins=50, alpha=0.8,
    color='#F44336',
    ax=axes[1],
    label='Fraud'
)
axes[1].set_title('Transaction Amount: Legit vs Fraud', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Amount (€)')
axes[1].set_ylabel('Count')
axes[1].legend()

# Save the chart to reports folder
plt.tight_layout()
plt.savefig('../reports/class_distribution.png', dpi=150, bbox_inches='tight')
plt.show()

print("Chart saved to reports/class_distribution.png")
```

```python
# Find which features (V1-V28, Amount, Time) are most related to fraud
# Correlation ranges from -1 to 1
# Close to 1 = when this feature goes up, fraud goes up
correlations = df.corr()['Class'].drop('Class').sort_values()

# Plot horizontal bar chart
plt.figure(figsize=(10, 12))

# Color bars — negative correlation = red, positive = green
colors = ['#F44336' if x < 0 else '#4CAF50' for x in correlations]

correlations.plot(kind='barh', color=colors, figsize=(10, 12))

plt.title('Feature Correlation with Fraud', fontsize=14, fontweight='bold')
plt.xlabel('Pearson Correlation')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')

plt.tight_layout()
plt.savefig('../reports/correlation_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()

# Print top features most related to fraud
print("TOP 5 POSITIVELY CORRELATED WITH FRAUD")
print(correlations.tail(5))

print(" TOP 5 NEGATIVELY CORRELATED WITH FRAUD")
print(correlations.head(5))
```

```python
# Compare transaction amounts between legitimate and fraud
# Do fraudsters spend more or less than legitimate users?

fig,axes=plt.subplots(1,2,figsize=(14, 5))

legit=df[df['Class']==0]
fraud=df[df["Class"]==1]

# ── Chart 1: Amount distribution ───
# Shows how much money is involved in legit vs fraud
legit["Amount"].hist(
    bins=50,alpha=0.6,
    color='#2196F3',
    ax=axes[0],
    label='Legitimate'

)
fraud["Amount"].hist(
    bins=50,alpha=0.6,
    color='#F44336',
    ax=axes[0],
    label='Fraud'
)
axes[0].set_title('Transaction Amount Distribution', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Amount (€)')
axes[0].set_ylabel('Count')
axes[0].set_yscale('log')  # log scale so fraud is visible
axes[0].legend()


# ── Chart 2: Average amount comparison ────────────────────
# Simple bar chart comparing average transaction amount
avg_amounts = [legit['Amount'].mean(), fraud['Amount'].mean()]
axes[1].bar(
    ['Legitimate', 'Fraud'],
    avg_amounts,
    color=['#2196F3', '#F44336'],
    edgecolor='white'
)
axes[1].set_title('Average Transaction Amount', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Average Amount (€)')


plt.tight_layout()
plt.savefig('../reports/amount_analysis.png', dpi=150, bbox_inches='tight')
plt.show()

# Print stats
print("LEGITIMATE TRANSACTIONS")
print(legit['Amount'].describe().round(2))

print("\nFRAUD TRANSACTIONS")
print(fraud['Amount'].describe().round(2))
```

```python
# Analyse when fraud happens — which hours of the day?
# Time column is in seconds — we convert to hours to make it readable
# Dataset covers 48 hours so we use modulo 24 to get hour of day

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Convert seconds to hours
# % 24 gives us hour of day (0-23) instead of total hours (0-48)
legit['Hour'] = (legit['Time'] / 3600) % 24
fraud['Hour'] = (fraud['Time'] / 3600) % 24

# ── Chart 1: Transaction count by hour ────────────────────
axes[0].hist(legit['Hour'], bins=24, alpha=0.6,color='#2196F3', label='Legitimate')
axes[0].hist(fraud['Hour'], bins=24, alpha=0.8,color='#F44336', label='Fraud')
axes[0].set_title('When Do Transactions Happen?', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Hour of Day (0-23)')
axes[0].set_ylabel('Count')
axes[0].set_yscale('log')
axes[0].legend()

# ── Chart 2: Fraud count by hour ──────────────────────────
# Shows specifically which hours have most fraud
fraud_hours = fraud['Hour'].round().astype(int)
fraud_hours.value_counts().sort_index().plot(
    kind='bar',
    ax=axes[1],
    color='#F44336',
    edgecolor='white'
)
axes[1].set_title('Fraud Count by Hour of Day', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Hour of Day')
axes[1].set_ylabel('Fraud Count')

plt.tight_layout()
plt.savefig('../reports/time_analysis.png', dpi=150, bbox_inches='tight')
plt.show()

print("=== FRAUD BY HOUR ===")
print(fraud['Hour'].value_counts().sort_index())
```
